```python
"""导入用到的包"""
import pandas as pd
import jieba
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.externals import joblib

import warnings
warnings.filterwarnings('ignore')
```

```python
"""读入数据集"""
df_news = pd.read_excel('Dataset/train.xlsx', index_col=False, names=['data','label'],encoding='utf-8')
df_news['data'] = df_news['data'].astype(str)
```

```python
"""打印样本个数并查看"""
print("样本的个数： ", df_news.shape)
df_news.head()
```

```python
print(df_news['label'].value_counts())
```

```python
content = df_news.data.values.tolist()
```

```python
content
```

```python
content_S = []
for line in content:
    current_segment = jieba.lcut(line)
    if len(current_segment) >= 1 and current_segment != '\r\n': #换行符
        content_S.append(current_segment)
df_content=pd.DataFrame({'data_S':content_S})
```

```python
df_content.head(10)
```

```python
stopwords=pd.read_csv("stopwords.txt",index_col=False,sep="\n",quoting=3,names=['stopword'], encoding='utf-8')
```

```python
def drop_stopwords(contents,stopwords):
    contents_clean = []
    all_words = []
    for line in contents:
        line_clean = []
        for word in line:
            if word in stopwords or word == ' ':
                continue
            line_clean.append(word)
            all_words.append(str(word))
            
        if line_clean is None:
            line_clean.append("空")
        contents_clean.append(line_clean)
    return contents_clean,all_words
```

```python
contents = df_content.data_S.values.tolist()  
```

```python
stopwords = stopwords.stopword.values.tolist()
data_clean,all_words = drop_stopwords(contents,stopwords)
df_content=pd.DataFrame({'data_clean':data_clean})
df_train=pd.DataFrame({'data_clean':data_clean,'label':df_news['label']})
```

```python
df_train.head(30)
```

```python
for d, l in df_train.iterrows():
    l.label = l.label.replace('[','').replace(']','').replace("'",'').split(',')
    print(l.label)
```

```python
df_train.head(30)
```

```python
for d, l in df_train.iterrows():
    if(len(l.label)>1):
        for item in l.label:
            temp = pd.DataFrame({"data_clean":[df_train.iloc[d].data_clean], "label":item.strip()})
            df_train = pd.concat([df_train, temp], axis=0)
        df_train.drop(d, inplace=True)
```

```python
df_train.shape
```

```python
df_train.reset_index(inplace=True)
```

```python
df_train.head(10)
df_train.drop('index', axis=1, inplace=True)
```

```python
for i,v in df_train['label'].items():
    if (type(v) == list):
        df_train['label'][i] = v[0]
```

```python
df_train['label']
```

```python
df_train['label'].value_counts()
```

```python
label_mapping = {"None": 1, "Happiness": 2, "Sadness": 3, "Surprise": 4, "Fear": 5, "Anger": 6}
df_train['label'] = df_train['label'].map(label_mapping)
```

```python
x_train = df_train['data_clean']
y_train = df_train['label']
```

```python
print("X_train_shape: ", x_train.shape)
print("Y_train_shape: ", y_train.shape)
```

```python
words = []
for line_index in range(len(x_train)):
    try:
        #x_train[line_index][word_index] = str(x_train[line_index][word_index])
        words.append(' '.join(x_train[line_index]))
    except:
        print (line_index)
print (len(words))
print(words[0])
# vec = CountVectorizer(analyzer='word', max_features=4000, lowercase = False)
# countresult = vec.fit(words)
# x = vec.transform(words)
tfid_vec = TfidfVectorizer()
x = tfid_vec.fit_transform(words)
print(x.shape)
```

```python
print(x)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y_train, test_size=0.2, random_state=8)
```

```python
classifier = MultinomialNB()
result = classifier.fit(X_train, Y_train)
```

```python
np.exp(classifier.class_log_prior_)   # 每个属性的先验概率
```

```python
y_predict = classifier.predict(X_test)
```

```python
print('The accuracy of Navie Bayes Classifier is',classifier.score(X_test,Y_test))
print(classification_report(Y_test,y_predict))
```

```python
result = pd.DataFrame({"Test": X_test, "predict": y_predict})
```
